# Refining the data

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
albums_df = spark.read.parquet("../Trusted/albums.parquet")
tracks_df = spark.read.parquet("../Trusted/tracks.parquet")

tracks_df.toPandas()
```

```python
albums_df.toPandas()
```

```python
# Removing useless columns from tracks

tracks_df = tracks_df.drop("artists", "available_markets", "external_urls", "href", "is_local", "preview_url", "uri", "type")

tracks_df.toPandas()
```

```python
# Removing useless columns from albums

albums_df = albums_df.drop("album_group", "artists", "available_markets", "external_urls", "href", "images", "release_date_precision", "type", "uri")

albums_df.toPandas()
```

```python
# Renaming some columns

albums_df = albums_df.withColumnsRenamed({"id": "album_id", "name": "album_name"})
tracks_df = tracks_df.withColumnsRenamed({"id": "track_id", "name": "track_name"})

albums_df.toPandas()
```

### Saving the data

```python
albums_df.write.parquet("..)
```
