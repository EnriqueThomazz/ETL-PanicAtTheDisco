# Transforming the data

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
# Loading the raw data

albums_df = spark.read.option("multiline", "true").json("../Raw/albums.json")
tracks_df = spark.read.option("multiline", "true").json("../Raw/tracks.json")

albums_df.toPandas()
```

```python
tracks_df.toPandas()
```

### Handling erros

```python
# Removing duplicates

albums_df = albums_df.drop_duplicates()

tracks_df = tracks_df.drop_duplicates()
```

```python
# Checking if there are any null values in tracks. For albums, its easy to see that there arent any null values (only 20ish values)
from pyspark.sql import functions as F
from functools import reduce

# I was seeing that the column preview_url is always null for every row, so Im not taking it into account
columns = [F.col(c) for c in tracks_df.drop("preview_url").columns]

null_rows = tracks_df.filter(reduce(lambda x, y: x | y.isNull(), columns[1:], columns[0].isNull()))

null_rows.toPandas()
```

### Enriching the data

```python
# Adding a column to albums with their total length

albums_df.createOrReplaceTempView("albums_df")
tracks_df.createOrReplaceTempView("tracks_df")

albums_df = spark.sql("""
SELECT *
FROM albums_df a LEFT JOIN (
    SELECT a.id as temp_id, sum(t.duration_ms) as total_duration_ms
    FROM albums_df a LEFT JOIN tracks_df t on a.id = t.album_id
    GROUP BY a.id
    ) b on a.id = b.temp_id
""")

albums_df = albums_df.drop("temp_id")

albums_df.toPandas()
```

### Saving the data

```python
# Saving to parquet files

albums_df.write.parquet("../Trusted/albums.parquet", mode="overwrite")
tracks_df.write.parquet("../Trusted/tracks.parquet", mode="overwrite")
```
